```python
from scipy.special import comb
# 正規分布関数の定義
def norm(x, mu, sigma):
    return np.exp(-((x - mu) / sigma) ** 2 / 2) / (np.sqrt(2 * np.pi) * sigma)

n = 1000

# グラフ
plt.xlim((430, 570))
x = np.arange(430, 571)

# 確率分布のグラフ
y1 = [comb(n, i) / 2 ** n for i in x]
plt.bar(x, y1)

# 正規分布関数のグラフ
mu = n / 2
sigma = np.sqrt(mu / 2)
y2 = norm(x, mu, sigma)
plt.plot(x, y2, c='k')

plt.title("確率分布と正規分布関数の重ね描き")
```
